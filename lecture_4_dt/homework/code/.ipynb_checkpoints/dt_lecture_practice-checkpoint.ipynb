### Задание 2. Обучение дерева решений (10 баллов)

Предобработайте признаки так, чтобы их можно было подать в модель дерева решений и обучите ее, постарайтесь подобрать оптимальные гиперпараметры, которые вам известны. Постарайтесь также обосновать выбор метрики в данном случае. При подборе гиперпараметров может помочь `GridSearchCV` [ссылка](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).

```python
# Удаление строк с пропущенными значениями
data_cleaned = data.dropna(axis=0)
data_cleaned.info()
```

```python
data_cleaned['key'].unique() #уникальные значения для key после удаления строк с пропущенными значениями
```

```python
#удалим неинформативные колонки
X = data_cleaned.drop(['Class', 'Artist Name', 'Track Name'],  axis=1)
y = data_cleaned['Class']

#сделаем шкалирование данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#разделим выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()

#зададим параметры для GridSearchCV

parametrs = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 6, 13]
}

grid_search = GridSearchCV(estimator=model, param_grid=parametrs, scoring='recall_weighted') #recall для многоклассовой классификации

#accuracy не подойдет, так как классы не сбалансированы
#recall максимизирует кол-во правильных ответов, weighted учитывает вес каждого класса, пропорциональный размеру класса 

grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_score_) 
```

```python
y_pred = grid_search.predict(X_test)
y_pred
```
